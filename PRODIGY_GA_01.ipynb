```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
```

1. Convert the sentences into the tokens

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2-large')
```

```python
model = GPT2LMHeadModel.from_pretrained('gpt2-large', pad_token_id=tokenizer.eos_token_id)
```

```python
tokenizer.eos_token_id
```

```python
tokenizer.decode(tokenizer.eos_token_id)
```

```python
sentence = 'I love travelling the world'
numeric_ids = tokenizer.encode(sentence, return_tensors='pt')

```

```python
numeric_ids
```

```python
tokenizer.decode(numeric_ids[0][2])
```

2. Generate the text given the sentence

```python
result = model.generate(numeric_ids,max_length=1000, num_beams=5,no_repeat_ngram_size=5, early_stopping=True)
```

```python
result
```

```python
generated_text = tokenizer.decode(result[0], skip_special_token=True)
print(generated_text)
```
